```python
import twitter
from twitter_settings import CONSUMER_KEY, CONSUMER_SECRET, OAUTH_TOKEN, OAUTH_TOKEN_SECRET

# CONSUMER_KEY = ''
# CONSUMER_SECRET = ''
# OAUTH_TOKEN = ''
# OAUTH_TOKEN_SECRET = ''

auth = twitter.oauth.OAuth(OAUTH_TOKEN, OAUTH_TOKEN_SECRET,
                           CONSUMER_KEY, CONSUMER_SECRET)

twitter_api = twitter.Twitter(auth=auth)

print twitter_api
```

```python
class User:
    name = ""
    id = 0
    protected = False
    friends = []
    def __init__(self, id):
        self.id = id
```

```python
def populate_user_name(user_list):
    name_dict = {}
    limit = 100
    i = 0
    j = 0
    list_len=len(user_list)
    while j < list_len:
        i = j
        j += limit
        list_slice = user_list[i:min(j,list_len)]
        keys_str = ",".join(map(lambda x: str(x.id), list_slice))#map to convert each element
        response = twitter_api.users.lookup(user_id = keys_str)
        for i in range(len(list_slice)):
            list_slice[i].name = response[i]["name"]
            list_slice[i].protected = response[i]["protected"]

# len(get_name_by_id(ischooler_list[:101]))
```

```python
def intersect(a, b):
     return list(set(a) & set(b))
    
def populate_friends(user_list, group_ids, limit=350):
    for user in user_list:
        if user.protected == False:
            user.friends = intersect(twitter_api.friends.ids(user_id = user.id)['ids'],group_ids)
        else:
            user.friends = []
        time.sleep(120)
    
```

```python
ischooler_ids = twitter_api.friends.ids(screen_name="BerkeleyISchool", count=350)['ids']
# ischooler_ids = ["27646969", "2439209120"] 
 
```

```python
ischooler_list = [User(x) for x in ischooler_ids]    
populate_user_name(ischooler_list)
```

```python
populate_friends(ischooler_list[45:350], ischooler_ids)
```

```python
with open("twitter_friend.json", "w") as json_file:
    json.dump(ischooler_list, json_file, default=lambda o: o.__dict__, sort_keys=True, indent=4)
```
