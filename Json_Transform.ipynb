```python
import json
inputfile=open('user_id_name_mapping.txt', 'r')
user_id_name_mapping = json.load(inputfile)
inputfile.close()

def find_id(name):
    for key in user_id_name_mapping.keys():
        if user_id_name_mapping[key] == name:
            return key
    return -1
```
